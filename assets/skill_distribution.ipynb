```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data - replace with your actual extracted skills count
skills_count = {'Python': 5, 'SQL': 4, 'NLP': 3, 'Java': 2, 'TensorFlow': 3}
keys = list(skills_count.keys())
vals = list(skills_count.values())

plt.figure(figsize=(10, 6))
sns.barplot(x=keys, y=vals, palette='viridis')
plt.title('Common Skills Identified Across Resumes', fontsize=14)
plt.xlabel('Skills', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save to assets
plt.savefig('../assets/skill_distribution.png', bbox_inches='tight', dpi=300)
plt.show()
```
