```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

class SimilarityEngine:
    def __init__(self):
        self.vectorizer = TfidfVectorizer()

    def calculate_score(self, jd_text, resume_texts):
        """Calculates cosine similarity between a JD and multiple resumes."""
        all_content = [jd_text] + resume_texts
        tfidf_matrix = self.vectorizer.fit_transform(all_content)
        
        # Compare first document (JD) with the rest (Resumes)
        scores = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:])
        return scores[0]
```
